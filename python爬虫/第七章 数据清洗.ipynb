```python
## 
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
import string
from collections import OrderedDict

def cleanInput(Inputs):
    Inputs = re.sub('\n+', ' ', Inputs)
    Inputs = re.sub(' +', ' ', Inputs)
    Inputs = re.sub('\[[0-9]*\]', "", Inputs)
    Inputs = bytes(Inputs, 'utf-8')
    Inputs = Inputs.decode('ascii', 'ignore')
    Inputs = Inputs.split(' ')
    clean = []
    for item in Inputs:
        item = item.strip(string.punctuation)
        if len(item) > 1 or (item.lower() == 'a' or item.lower() == 'i'):
            clean.append(item)
    return clean
    
def ngrams(Inputs, n):
    Inputs = cleanInput(Inputs)
    outputs = []
    for i in range(len(Inputs) - n - 1):
        outputs.append(Inputs[i:i + n])
    return outputs

html = urlopen("http://en.wikipedia.org/wiki/Python_(programming_language)")
bsObj = BeautifulSoup(html)
content = bsObj.find("div", {"id":"mw-content-text"}).get_text()
ng = ngrams(content, 2)
ng = OrderedDict(sorted(ng, key=lambda t: t[1], reverse=True))
print(ng)
```
